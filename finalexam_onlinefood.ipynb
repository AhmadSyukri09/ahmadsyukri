# Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
food = pd.read_csv("onlinefoods.csv")
food
```

```python
#variabel selection
df= food[['Age', 'Gender', 'Occupation', 'MonthlyIncome', 'EducationalQualifications','Feedback']]
df.head()
```

```python
# Display basic information about the dataset
print(df.info())
print(df.describe())
```

```python
# Drop the 'Unnamed: 12' column as it is a duplicate
df = df.drop(columns=['Unnamed: 12'])
```

```python
df
```

# Handle Missing Value


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Handle missing values
df = df.fillna(df.mode().iloc[0])
df.info()
```

# Mengkodekan Variabel Kategori

```python
# Encode categorical variables
categorical_cols = ['Gender', 'Occupation', 'EducationalQualifications']
label_encoders = {col: LabelEncoder() for col in categorical_cols}

def encode_MonthlyIncome(MonthlyIncome):
    if MonthlyIncome == 'No Income':
        return 0
    elif MonthlyIncome == 'Below Rs.10000':
        return 1
    elif MonthlyIncome == '10001 to 25000':
        return 2
    elif MonthlyIncome == '25001 to 50000':
        return 3
    elif MonthlyIncome == 'More than 50000':
        return 4
    else:
        return None

# Terapkan fungsi pada kolom 'Income'
df['MonthlyIncome'] = df['MonthlyIncome'].apply(encode_MonthlyIncome)

for col in categorical_cols:
    df[col] = label_encoders[col].fit_transform(df[col])

# Convert 'Output' to binary: Yes=1, No=0
df['Feedback'] = df['Feedback'].apply(lambda x: 1 if x == 'Positive' else 0)
```

```python
df
```

```python
# Scale numerical features

# Split the dataset into training and testing sets
x = df[['Age', 'Gender', 'Occupation', 'MonthlyIncome', 'EducationalQualifications']]
y = LabelEncoder().fit_transform(df['Feedback'])
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# EDA Visualization
sns.countplot(x='Feedback', data=df)
plt.title('Feedback Distribution')
plt.show()

sns.boxplot(x='Feedback', y='Age', data=df)
plt.title('Age vs Feedback')
plt.show()

sns.countplot(x='MonthlyIncome', hue='Feedback', data=df)
plt.title('Monthly Income vs Output')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Model Training and Comparison
def evaluate_model(y_test, y_pred):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    return accuracy, precision, recall, f1

# Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

# SVM
svm = SVC()
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)

# Evaluate models
metrics_lr = evaluate_model(y_test, y_pred_lr)
metrics_dt = evaluate_model(y_test, y_pred_dt)
metrics_svm = evaluate_model(y_test, y_pred_svm)

print("Logistic Regression - Accuracy: {:.2f}, Precision: {:.2f}, Recall: {:.2f}, F1-Score: {:.2f}".format(*metrics_lr))
print("Decision Tree - Accuracy: {:.2f}, Precision: {:.2f}, Recall: {:.2f}, F1-Score: {:.2f}".format(*metrics_dt))
print("SVM - Accuracy: {:.2f}, Precision: {:.2f}, Recall: {:.2f}, F1-Score: {:.2f}".format(*metrics_svm))
```
